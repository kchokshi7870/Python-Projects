$INTRODUCTION$

In the burgeoning field of optoelectronics, Deep UV Micro-Light-Emitting Diodes (DUV-μLEDs), particularly those fabricated from Aluminum Gallium Nitride (AlGaN), represent a critical advancement with profound implications across a spectrum of applications. Notably utilized in sterilization, UV phototherapy, and non-line-of-sight communications, these devices offer a promising alternative to traditional mercury lamps, aligning with global efforts toward environmental sustainability and enhanced public health safety.

Despite their potential, the performance of DUV-μLEDs is heavily influenced by their physical dimensions, which affect their optical power output, efficiency, and operational bandwidth. The importance of size in these devices is underscored by their surface-to-volume ratio, which impacts thermal management and efficiency, critical factors in their practical applications. However, the literature indicates a gap in comprehensive studies focused specifically on the size-dependent characteristics of these LEDs, particularly in how these dimensions influence their efficiency and application potential.

This project seeks to address this gap by examining the correlation between the size of DUV-μLEDs and their performance attributes. Through systematic experimentation and analysis, this study will explore whether smaller or larger DUV-μLEDs exhibit superior performance metrics such as power output and bandwidth, providing insights that could drive future innovations in LED technology.

This investigation leverages a dataset meticulously compiled in prior research conducted by Maclure et al., as part of the 2020 IEEE Photonics Conference. In their groundbreaking work, "Size-dependent characterisation of deep UV micro-light-emitting diodes," they employed a Thorlabs PM100A power meter to capture the spectral power output, and a Yokogawa GS610 Source Measure Unit to record the corresponding currents at varying dimensions of DUV-μLEDs. The systematic collection of this data, under controlled experimental setups, ensures the reliability and accuracy of the measurements, providing a robust basis for further analysis. By re-examining this data, this project aims to derive significant insights into the size-dependent behaviors of DUV-μLEDs, potentially uncovering new relationships and optimizing performance parameters that were not fully explored in the original study.

The outcomes of this investigation are anticipated to contribute not only to the academic understanding of DUV-μLEDs but also to the practical enhancement of their design and application, paving the way for more efficient and effective UV solutions in various technological and medical fields.

```python
# Hypothesis testing

# In our analysis of the dataset concerning DUV-μLEDs, we employed specific hypothesis testing to rigorously evaluate 
# our best-fit models. The Overall Model Significance Test, an F-test, was chosen to determine the statistical 
# significance of the model as a whole compared to a baseline model with no predictors. This was essential to 
# validate that the model's predictors meaningfully improve the prediction of power output. We also used an F-test 
# to compare nested models, allowing us to test if the incorporation of a quadratic term significantly enhanced the 
# model fit compared to a simpler linear model and the comparision between cubic and quadratic terms. 
# These tests are particularly crucial as we aim to compare different parts of our dataset, discerning the impacts 
# of varying LED sizes on power output when subjected to different currents. Notably, the quadratic model's fitness 
# was confirmed to be identical to the results obtained from the np.polyfit method, providing reassurance of the 
# robustness and accuracy of our regression analysis. 
# F-tests assess the need for increasingly complex polynomial models for DUV-μLEDs. Each F-test examines the null 
# hypothesis (H0) that a simpler model, such as linear over quadratic, is sufficient. The alternative hypothesis (H1) 
# contends that a more complex model significantly enhances the fit. A p-value less than 0.05 typically rejects H0, 
# indicating that the added complexity of the model—quadratic or cubic—improves the fit significantly.

# hypothesis test for size 20 micro-meters
import numpy as np
import statsmodels.api as sm
x=np.array(current1)
y=np.array(power1)
# Prepare the design matrices for the linear, quadratic, and cubic models
x_lin = sm.add_constant(x)  # Linear
x_quad = sm.add_constant(np.column_stack((x, x**2)))  # Quadratic
x_cubic = sm.add_constant(np.column_stack((x, x**2, x**3)))  # Cubic

# Fit the models
model_lin = sm.OLS(y, x_lin).fit()
model_quad = sm.OLS(y, x_quad).fit()
model_cubic = sm.OLS(y, x_cubic).fit()

# Overall Model Significance Test for the quadratic model (compared to no predictors)
print("Overall Model Significance Test for the quadratic model:")
print(model_quad.f_test(np.identity(len(x_quad[0]))[:-1]))  # Test all terms except the constant

# Nested Model Comparison: Linear vs Quadratic
print("\nComparison of linear to quadratic model:")
print(model_quad.compare_f_test(model_lin))

# Nested Model Comparison: Quadratic vs Cubic
print("\nComparison of quadratic to cubic model:")
print(model_cubic.compare_f_test(model_quad))
```

$DISCUSSION$

Our investigation into the size-dependent power output characteristics of Deep UV Micro-Light-Emitting Diodes (DUV-μLEDs) corroborates the hypothesis that LED size significantly influences power output under variable current conditions. The empirical data, delineated by the distinct upward trends in the current vs. power graphs, proves the proposition that larger DUV-μLEDs exhibit increased power output, which is true due to a larger active area which presumably supports the generation of a greater number of photons. This phenomenon aligns with photonics principles which assert that larger emissive areas in LEDs should theoretically yield higher photon outputs, given a proportionate current supply. However, a comprehensive comparison with extant literature is hampered by a scarcity of specific studies on size-dependent efficiencies in DUV-μLEDs, a niche yet burgeoning field of photonic research. 

The results from the Overall Model Significance Tests for the quadratic model provide compelling evidence that the quadratic terms are statistically significant across all sizes of Deep UV Micro-Light-Emitting Diodes (DUV-μLEDs). Notably, the F-statistics (ranging from 400.96 for the smallest size to 46,979.28 for the largest size) and the corresponding p-values (ranging from 2.95e-42 for the largest size to 2.83e-10 for the smallest size) demonstrate an overwhelmingly significant model fit, suggesting that the quadratic relationship between current and power output robustly applies across different sizes of LEDs. Intriguingly, the trend of increasing F-statistics with LED size might indicate that the quadratic relationship becomes more pronounced in larger LEDs, possibly due to enhanced physical characteristics such as a larger active area, which facilitates greater electron-hole recombination efficiency. These results affirm the critical role of LED size in influencing the efficacy and performance characteristics of DUV-μLEDs, with larger LEDs not only adhering to expected theoretical models but also showing a stronger quadratic dependency in power output response. This could be attributed to the increased complexity and dynamics in photon generation within larger semiconductive structures.

The comparative analysis of linear versus quadratic models and quadratic versus cubic models for DUV-μLEDs reveals a consistent pattern across varying sizes. The F-tests consistently show that the quadratic model provides a significantly better fit than the linear model across all sizes, with F-statistics ranging from 73.14 to 8906.51 and p-values diminishing into the range of 1.38e-13 to 2.77e-31. These results underscore the inadequacy of the linear model in capturing the complexities of the power output as current varies, affirming the hypothesis that the power output's relationship with current is inherently non-linear, particularly pronounced as the size of the LEDs increases. Conversely, the transitions from quadratic to cubic models exhibit a less consistent pattern. While for the smaller sizes, the addition of a cubic term did not result in statistically significant improvements (p-values ranging from 0.119 to 0.706), the largest size demonstrated a significant improvement with a p-value of 
0.000858, suggesting that for larger LEDs, a cubic relationship might be emerging, possibly due to more complex physical interactions within the LED structure as size increases. This variation in model enhancements from quadratic to cubic highlights potential size-dependent non-linear dynamics that could be critical in optimizing DUV-μLED designs for specific applications. Future studies should delve deeper into these size-specific behaviors, potentially integrating more complex or even higher-dimensional polynomial models to refine the understanding of these dynamics.

Our regression analysis, besides providing p-values indicating significant model improvements, also calculated 95% confidence intervals for each coefficient. These intervals not only underscore the statistical significance of the coefficients (as they do not overlap with zero) but also offer a quantifiable measure of estimation uncertainty. For example, the confidence interval for the quadratic term widened with increasing LED size, suggesting greater variability in its effect on power output among larger LEDs. Such insights are crucial for understanding the precision of our estimates and the reliability of our conclusions under varying experimental conditions.

The observed results are logically consistent with established photonic behaviors and material properties of LEDs. Potential sources of experimental error, including but not limited to variations in material quality and measurement inaccuracies, could be avoided in future studies by employing more uniform fabrication processes and advanced metrological tools to make the LEDs. Further research, with a broader array of LED sizes and more granular variations in current, could refine these initial observations and foster a deeper understanding of the size-related dynamics in DUV-μLED power output.

In conclusion, our analysis demonstrated that the size of DUV-μLEDs significantly influences their power output response to varying currents, with larger LEDs exhibiting more pronounced increases. The F-tests confirmed that quadratic and cubic models significantly enhance the predictive accuracy over simpler linear models, particularly as LED size increases. The p-values obtained strongly supported the rejection of simpler models in favor of more complex ones, suggesting a nuanced interplay between LED size and power output dynamics. Furthermore, the confidence intervals provided insights into the precision of our estimates, highlighting the rigor of our findings. These results not only validate our initial hypotheses but also underscore the importance of considering LED size in the design and utilization of DUV-μLEDs for optimal performance.
